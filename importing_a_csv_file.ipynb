## mods we need

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## import data


```python
SP = pd.read_csv("~/Desktop/s&p 500 data/^GSPC.csv", index_col = "Date",
                parse_dates=True, na_values=['nan'])
```

```python
print (SP.head())
```

## funtion for plotting data


```python
def plot_data(A, title = "S&P Data"):
    ax = A.plot (fontsize=12)
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    plt.show()
```

## plotting data

```python
plot_data(SP['Close'])
```

## daily returns

```python
 def compute_daily_returns(B):
        DR = B.copy()
        DR[1:] = (B[1:]/B[:-1].values)-1
        DR.ix[0, :] = 0
        return DR
        
```

```python
Daily_Returns = compute_daily_returns(SP)
```

```python
print (Daily_Returns.head())
```

```python
plot_data(Daily_Returns["Close"])
```

## mean and sd

```python
Mean = Daily_Returns['Close'].mean()
```

```python
print (Mean)
```

```python
std = Daily_Returns['Close'].std()
```

```python
print (std)
```

## histogram

```python
Daily_Returns['Close'].hist(bins=100)

plt.axvline(Mean, color = 'w' , linestyle = 'dashed',linewidth = 2)
plt.axvline(std, color = 'r' , linestyle = 'dashed',linewidth = 2)
plt.axvline(-std, color = 'r' , linestyle = 'dashed',linewidth = 2)
plt.show()
```
